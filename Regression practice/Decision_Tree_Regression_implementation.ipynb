```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("datasets/Position_Salaries.csv")
dataset
```

```python
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

-----------------------------------------

**no feature scaling is required in decision trees**

reason: the prediction from decision tree regression or random forest regression model are resulting from successive splits of the data, through different nodes of the tree. Therefore they are not some equations like other models. 

Thats why no feature scaling is needed to split the different values of feature into different categories leading to different predictions.

We can still do this with original scales of the features even if the features take different ranges of values.

---------------------------------------

Train the decision tree model on the whole dataset

we can also do train test split and train the model on train set.

I have not splitted the data because I want my model to be beneficial by training it with the whole dataset and I will predict by entering new data

```python
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state= 0)
model.fit(X, y)
```

predict a new result

```python
model.predict([[6.5]])
```

vizualise the decision tree regressor results (high resolution)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X, y, color='r')
plt.plot(X_grid, model.predict(X_grid), color='b')
plt.xlabel('Position-level')
plt.ylabel('Salary')
plt.show()
```
