```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv("datasets/Position_Salaries.csv")
dataset
```

```python
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

Train the Random Forest Regrssor model on whole dataset

we can also do train test split and train the model on train set.

I have not splitted the data because I want my model to be beneficial by training it with the whole dataset and I will predict by entering new data

```python
from sklearn.ensemble import RandomForestRegressor
model = RandomForestRegressor(n_estimators= 10, random_state= 0 )
model.fit(X,y)
```

predict a new result

```python
model.predict([[6.5]])
```

Vizualise the results of Random Forest Regressor (higher resolution)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape(len(X_grid),1)
plt.scatter(X, y, color='r')
plt.plot(X_grid, model.predict(X_grid), color='b')
plt.xlabel('Position-level')
plt.ylabel('Salary')
plt.show()
```

Here more steps between data points compared to decision tree because we have more trees here, and therefore more splits of the features where you have the same prediction, the avg of predicted salary.
