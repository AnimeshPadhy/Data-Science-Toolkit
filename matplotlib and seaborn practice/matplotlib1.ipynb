```python
import matplotlib.pyplot as plt
import numpy as np 
```

```python
%matplotlib inline
```

```python
x = np.linspace(1,10,15)
print(x)
y = x**2
print(y)
```

Functional Method

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

Object Oriented method

```python
fig = plt.figure()

               # left,bottom,width,height
axes = fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('title')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])
axes3 = fig.add_axes([0.5,0.15,0.4,0.3])
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.55,0.4,0.3])

axes1.plot(x,y,color='red',linewidth=3,alpha=0.5,marker='o')
axes1.set_title('larger plot')
axes2.plot(y,x,color='green',lw=0.8,linestyle='--')
axes2.set_title('smaller plot')
```
